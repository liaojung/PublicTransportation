**分隔島** （ 綜合資料 ）

----------------------------------------



```python
onboard = yilan_OD.groupby(['附屬路線唯一代碼','起站.站牌代碼','路線方向.0.去程.1.返程.']).sum(numeric_only = True).drop('旅次時段.Ex..12表示為12時00分至12時59分.', axis = 1).rename(columns = {'載客量':'上車人數'})
onboard = onboard.reset_index(names =['附屬路線唯一代碼','站牌代碼','路線方向'])

display(onboard['上車人數'].sum())
onboard
```

```python
offboard = yilan_OD.groupby(['附屬路線唯一代碼','迄站.站牌代碼','路線方向.0.去程.1.返程.']).sum(numeric_only = True).drop('旅次時段.Ex..12表示為12時00分至12時59分.', axis = 1).rename(columns = {'載客量':'下車人數'})
offboard = offboard.reset_index(names =['附屬路線唯一代碼','站牌代碼','路線方向'])

display(offboard['下車人數'].sum())
offboard
```

```python
data = Yilan_bus_station.merge(onboard, how='left', left_on=['SubRouteUID','StopUID','Direction'], right_on=['附屬路線唯一代碼','站牌代碼','路線方向']).drop(['附屬路線唯一代碼','站牌代碼','路線方向'],axis = 1)
data = data.merge(offboard, how='left', left_on=['SubRouteUID','StopUID','Direction'], right_on=['附屬路線唯一代碼','站牌代碼','路線方向']).drop(['附屬路線唯一代碼','站牌代碼','路線方向'],axis = 1)

data = data.fillna(0)
data = data.astype({'上車人數': 'int64', '下車人數': 'int64'})

print(data['下車人數'].sum())
print(data['上車人數'].sum())
data
```
